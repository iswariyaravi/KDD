```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/Users/iswariyaravi/Downloads/IBM_Attrition_v3 (1).csv")
df
```

# **a)	Delete all the rows with missing value.**

```python
df = df.dropna(axis=0)
df
```

# **b)	Create four categories (income1, income2, income3, income4) based on monthly income (“MonthlyIncome”) as:**

```python
def income_groups(x):
    if x <= 2900:
        return 'income1'
    elif 2900 < x <= 5000:
        return 'income2'
    elif 5000< x <= 8500 :
        return 'income3'
    else:
        return 'income4'

df['MonthlyIncome'] = df['MonthlyIncome'].apply(income_groups)

df
```

# **c)	Create two categories (senior, not-senior) for years at the company (“YearsAtCompany”):**

```python
df['YearsAtCompany'] = df['YearsAtCompany'].apply(lambda x : 'Senior' if x<=6 else 'not senior')
df
```

# **d)	Create two categories (young, mature) for age:**



```python
df['Age'] = df['Age'].apply(lambda x: 'young' if x<=37 else 'mature')
df
```

# problem 2 creating test and train set and applying random forest

```python
# print(dtype(df['Age']))

df['Age'].dtypes
```

```python
#label encoding

from sklearn import preprocessing
le = preprocessing.LabelEncoder()
for i in df.columns:
    if df[i].dtypes =='O':
        df[i]= le.fit_transform(df[i])

df
```

```python
test = df[df.index % 4 == 0]
test
test.reset_index(drop = True)
```

```python
train = df[df.index % 4 != 0] 
train.reset_index(drop = True)
train
```

```python
from sklearn.ensemble import RandomForestClassifier

import sklearn

clf = RandomForestClassifier(max_depth=5,
                            , random_state=0)
clf.fit(train.iloc[:, :-1], train.iloc[:, -1])

ypred = clf.predict(test.iloc[:, :-1])


accuracy = sklearn.metrics.accuracy_score(test.iloc[:, -1], ypred)
print('ACCURACY OF RANDOM FOREST:' , accuracy)
```

# problem 3 implementing decision tree c5.0

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=2702, criterion ='entropy', min_samples_split =5)

clf.fit(train.iloc[:, :-1], train.iloc[:, -1])

ypred = clf.predict(test.iloc[:, :-1])


accuracy = sklearn.metrics.accuracy_score(test.iloc[:, -1], ypred)
print('ACCURACY OF Decision tree:' , accuracy)
```
